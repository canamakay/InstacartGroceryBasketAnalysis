### Step 4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#First we must create a 'max_order' column
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking to see whether 'max_order" columnn has been added 
df_ords_prods_merge.head()
```

```python
#creating a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value.
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking to see if loyalty flag colummn has been added 
df_ords_prods_merge.head()
```

```python
#checking the frequency of values in the loyalty column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

### Step 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#first we must make a new column for average amount spent 
df_ords_prods_merge['average_spent'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#checking to see if average_spent column has been added
df_ords_prods_merge.head()
```

```python
#creating a spending flag column
df_ords_prods_merge.loc[df_ords_prods_merge['average_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#checking to see whether spending_flag colummn has been added
df_ords_prods_merge.head()
```

```python
#checking the frequency of values of the spending flag column
df_ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

### Step 7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer."
### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer ”
### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#first we must create a new column for median_days_since_last_order
df_ords_prods_merge['median_days_since_last_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
#checking to see if median_days_since_last_order column has been added
df_ords_prods_merge.head()
```

```python
#creating frequency flag column
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_last_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_last_order'] <= 20) & (df_ords_prods_merge['median_days_since_last_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_last_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#checking to see if frequency flag column has been added
df_ords_prods_merge.head()
```

```python
#checking the values for the frequency_flag column
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
